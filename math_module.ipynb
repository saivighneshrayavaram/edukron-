29.Create a grading system using comparison and logical operators.

```python
x=int(input("Enter the total subjects: "))
total=100*x
sum=0
for i in range(x):
  y=int(input("Enter Marks:" ))
  sum=sum+y
gain=((sum/total)*100)
if(gain >90):
  print("A")
elif(gain >80 and gain <=90):
  print("B")
elif(gain >70 and gain <=80):
  print("C")
elif(gain >60 and gain <=70):
  print("D")
elif(gain > 50 and gain <=60):
  print("E")
elif(gain <= 50):
  print("F")
```
